```python
from keras.models import Sequential
from keras.layers import Dense

import numpy as np
import tensorflow as tf
```

```python
pd.read_csv('ThoraricSurgery.csv',header=-1)
```

```python
# seed ë¥¼ í†µí•´ 0ë³´ë‹¤  í¬ê±°ë‚˜ ê°™ì€ ìž„ì˜ì˜ ì •ìˆ˜ë¥¼ ì¶”ì¶œ(ë™ì¼ seed ëŠ” ê°™ì€ ë‚œìˆ˜ê°€ ë°œìƒ )
seed =0
np.random.seed(seed)

tf.set_random_seed(seed)
Data_set= np.loadtxt('ThoraricSurgery.csv',delimiter=',')
```

```python
#ì „ì²´ ë°ì´í„° í–‰ ì¤‘ì—ì„œ 0ë²ˆì§¸ë¶€í„° 17ë²ˆì§¸ ì—´ ê°“ã…‚ë“¤ì„ xë¦¬ìŠ¤íŠ¸ì— ì €ìž¥
X=Data_set[:,0:17]
#ì „ì²´ ë°ì´í„°ì—ì„œ 17ë²ˆì§¸ ì—¬ê°’ì„ yë¦¬ìŠ¤íŠ¸ì— ì €ìž¥
y=Data_set[:,17]
```

```python
X
```

```python
#ì‹ ê²½ë§ì„ í•œì¸µ í•œì¸µ ìŒ“ëŠ” ê¸°ëŠ¥ (add()í•¨ìˆ˜ ì‚¬ìš© )
model=Sequential()
#Dense():ì‹ ê²½ë§ ê°ì¸µì˜ íŠ¹ì„±ì„ ì„¤ì •í•˜ëŠ” ê¸°ëŠ¥ 
model.add(Dense(30,input_dim=17,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
# compile í•™ìŠµ í”„ë¡œì„¸ìŠ¤ë¥¼ ì„¤ì •í•˜ëŠ” ê¸°ëŠ¥ 
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
model.fit(X,y,epochs=30,batch_size=10)
```

```python
#ê²°ê³¼ ì¶œë ¥

print ('\n ì •í™•ë„ :',model.evaluate(X,y)[1])
```
